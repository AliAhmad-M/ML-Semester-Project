### Loading Datasets

```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

print("Lahore Air Quality Index (AQI) Data from 2019 to 2023")
lahore_aqi_df = pd.read_csv("input/lahore_aqi_2019_to_2023.csv")
display(lahore_aqi_df.head())

print("\nPakistan Weather Data from 2000 to 2024")
pakistan_weather_df = pd.read_csv("input/pakistan_weather_2000_2024.csv")
display(pakistan_weather_df.head())

print("\nLahore Weather Data")
lahore_weather_df = pd.read_csv("input/lahore_weather_data.csv")
display(lahore_weather_df.head())
```

### Checking Columns

```python
print(f"Lahore AQI DataFrame Columns:")
for column in lahore_aqi_df.columns:
    print(column)

print(f"\nPakistan Weather DataFrame Columns:")
for column in pakistan_weather_df.columns:
    print(column)

print(f"\nLahore Weather DataFrame Columns:") 
for column in lahore_weather_df.columns:
    print(column)
```

### Filter Pakistan Weather Dataset to Lahore

```python
pakistan_weather_df = pakistan_weather_df[
    pakistan_weather_df['city'].str.lower() == 'lahore'
]

display(pakistan_weather_df.head())
```

### Checking Date Range

```python
# Lahore AQI
lahore_aqi_df['date'] = pd.to_datetime(lahore_aqi_df['date'], format="%d-%m-%y")
print("Lahore AQI Date Range:",
      lahore_aqi_df['date'].min().strftime("%d-%m-%y"),
      "to",
      lahore_aqi_df['date'].max().strftime("%d-%m-%y"))

# Pakistan Weather
pakistan_weather_df['date'] = pd.to_datetime(pakistan_weather_df['date'], format="%m/%d/%Y")
print("Pakistan Weather Date Range:",
      pakistan_weather_df['date'].min().strftime("%d-%m-%y"),
      "to",
      pakistan_weather_df['date'].max().strftime("%d-%m-%y"))

# Lahore Weather
lahore_weather_df['Date'] = pd.to_datetime(lahore_weather_df['Date'], format="%m/%d/%Y")
print("Lahore Weather Date Range:",
      lahore_weather_df['Date'].min().strftime("%d-%m-%y"),
      "to",
      lahore_weather_df['Date'].max().strftime("%d-%m-%y"))
```

### Standardize Date Columns

```python
lahore_weather_df.rename(columns={'Date': 'date'}, inplace=True)
```

### Check Overlapping Date Range

```python
start_date = max(lahore_aqi_df['date'].min(),
                 pakistan_weather_df['date'].min())

end_date   = min(lahore_aqi_df['date'].max(),
                 pakistan_weather_df['date'].max())

print("Overlapping date range:", start_date.strftime("%d-%m-%y"), 
      "to", end_date.strftime("%d-%m-%y"))
```

**Note:** We do not consider Lahore Weather Data, as it only contains data for first 2 days of each year

### Keep Only Overlapping Date Range

```python
lahore_aqi_df = lahore_aqi_df[(lahore_aqi_df['date'] >= start_date) & (lahore_aqi_df['date'] <= end_date)]
pakistan_weather_df = pakistan_weather_df[(pakistan_weather_df['date'] >= start_date) & (pakistan_weather_df['date'] <= end_date)]
lahore_weather_df = lahore_weather_df[(lahore_weather_df['date'] >= start_date) & (lahore_weather_df['date'] <= end_date)]
```

### Standardize All Column Names

```python
# Lahore AQI renaming
lahore_aqi_df = lahore_aqi_df.rename(columns={
    'max_temp_f': 'temp_max_c',
    'avg_temp_f': 'temp_avg_c',
    'min_temp_f': 'temp_min_c',
    'max_dew_point_f': 'dew_point_max_c',
    'avg_dew_point_f': 'dew_point_avg_c',
    'min_dew_point_f': 'dew_point_min_c',
    'max_humidity_percent': 'humidity_max_percent',
    'avg_humidity_percent': 'humidity_avg_percent',
    'min_humidity_percent': 'humidity_min_percent',
    'max_wind_speed_mph': 'wind_speed_max_mph',
    'avg_wind_speed_mph': 'wind_speed_avg_mph',
    'min_wind_speed_mph': 'wind_speed_min_mph',
    'max_pressure_in': 'pressure_max_in',
    'avg_pressure_in': 'pressure_avg_in',
    'min_pressure_in': 'pressure_min_in'
})

# Temperature columns in Lahore AQI that need conversion
temp_cols = ['temp_max_c', 'temp_avg_c', 'temp_min_c']
dew_cols  = ['dew_point_max_c', 'dew_point_avg_c', 'dew_point_min_c']

# Convert F -> C
for col in temp_cols + dew_cols:
    lahore_aqi_df[col] = (lahore_aqi_df[col] - 32) * 5 / 9

# Pakistan Weather renaming
pakistan_weather_df = pakistan_weather_df.rename(columns={
    'tmax': 'temp_max_c',
    'tmin': 'temp_min_c',
    'tavg': 'temp_avg_c',
    'dew_point': 'dew_point_c',
    'humidity': 'humidity_percent',
    'wspd': 'wind_speed_mph',
    'pressure': 'pressure_in',
    'prcp': 'precip_mm'
})

# Lahore Weather renaming
lahore_weather_df = lahore_weather_df.rename(columns={
    'temp_max (celcius)': 'temp_max_c',
    'temp_min (celcius)': 'temp_min_c',
    'preciptation (mm)': 'precip_mm',
    'soil_moisture (mm)': 'soil_moisture_mm'
})
```

### Merge All 3 Dataframes

```python
merged_df = (
    lahore_aqi_df
        .merge(pakistan_weather_df, on='date', how='left', suffixes=('', '_pak_weather'))
        .merge(lahore_weather_df, on='date', how='left', suffixes=('', '_lhr_weather'))
)

merged_df = merged_df.sort_values('date').reset_index(drop=True)
display(merged_df.head())
for column in merged_df.columns:
    print(column)
```

### Percentage of NULL Values Per Column

```python
null_percent = (merged_df.isnull().sum() / len(merged_df)) * 100
null_percent = null_percent.sort_values(ascending=False)
print(null_percent)
```

### Check Unique Values In Cagtegorical Columns

```python
# Unique values
print("Unique values in wind_category:")
print(merged_df["wind_category"].unique())

print("\nUnique values in rainfall_intensity:")
print(merged_df["rainfall_intensity"].unique())

print("\nUnique values in season:")
print(merged_df["season"].unique())
```

### Drop NULL and redundant columns

```python
# Drop visibility as it has 100% null values
# Drop redundant columns also
merged_df = merged_df.drop(columns=[
    'city', 'region', 'latitude', 'longitude', 'elevation', 'is_hot_day', 'is_cold_day',
    'temp_range', 'wind_category', 'soil_moisture_mm', 'visibility', 'region_lhr_weather'
])

for column in merged_df.columns:
    print(column)
```

- The columns - city, region, latitude, longitude, elevation remain same for Lahore
- The columns - visibility, soil_moisture_mm is 100% null
- The columns - temp_range, is_hot_day, is_cold_day can be determined using temperatures
- The column  - wind_category only has a single value, that is 'windy'

**Note:** We shall handle the precip_mm_lhr_weather, temp_max_c_lhr_weather, temp_min_c_lhr_weather columns later

### Remove Duplicate Columns

```python
# Same columns coming from multiple sources
merged_df = merged_df.drop(columns=[
    'temp_max_c_pak_weather', 'temp_avg_c_pak_weather', 'temp_min_c_pak_weather', 
    'temp_max_c_lhr_weather', 'temp_min_c_lhr_weather', 'precip_mm_lhr_weather',
    'dew_point_c', 'pressure_in', 'humidity_percent', 'wind_speed_mph'

])

for column in merged_df.columns:
    print(column)

print(f"\nTotal Number of Rows: {merged_df.shape[0]}")
```

### Impute NULL aqi_pm2.5

```python
from sklearn.impute import KNNImputer

# Ensure 2D array
aqi_values = merged_df[['aqi_pm2.5']].values

imputer = KNNImputer(n_neighbors=5)
merged_df['aqi_pm2.5'] = imputer.fit_transform(aqi_values)

print("Dataframe:")
display(merged_df.head())
print(f"\nNull Values:\n{merged_df.isnull().sum()}")
```

### Handle Outliers in Numerical Values

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = merged_df.select_dtypes(include=['float64', 'int64']).columns.drop(
    ['aqi_pm2.5', 'year', 'month', 'day', 'dayofweek', 'is_weekend']
)

plt.figure(figsize=(8, 16))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(len(numerical_cols)//3 + 1, 3, i)
    sns.boxplot(y=merged_df[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
# Define realistic bounds
temp_min, temp_max = -5, 55             # ceclius
wind_min, wind_max = 0, 80              # mph
pressure_min, pressure_max = 24, 36     # inHg
dew_point_min, dew_point_max = -5, 40   # celcius

# Filter rows within temperature, wind, pressure, and dew point bounds
merged_df = merged_df[
    (merged_df['temp_min_c'] >= temp_min) & (merged_df['temp_max_c'] <= temp_max) &
    (merged_df['wind_speed_min_mph'] >= wind_min) & (merged_df['wind_speed_max_mph'] <= wind_max) &
    (merged_df['pressure_min_in'] >= pressure_min) & (merged_df['pressure_max_in'] <= pressure_max) &
    (merged_df['dew_point_min_c'] >= dew_point_min) & (merged_df['dew_point_max_c'] <= dew_point_max)
]

# Check min and max after filtering
print(merged_df[['temp_min_c','temp_max_c','wind_speed_min_mph','wind_speed_max_mph',
                 'pressure_min_in','pressure_max_in','dew_point_min_c','dew_point_avg_c','dew_point_max_c']].agg(['min','max']))

print(f"Total rows after filtering extreme values: {merged_df.shape[0]}")
```

### Applying One-Hot Encoding to 'season' Column

```python
merged_df = pd.get_dummies(
    merged_df,
    columns=['season', 'rainfall_intensity'], prefix=['season', 'rain']
)

display(merged_df.head())
```

### Check the Correlation for Redundancy

```python
# Select rain related columns
rain_cols = [col for col in merged_df.columns if col.startswith("rain_")]
cols = ["precip_mm"] + rain_cols

df = merged_df[cols].apply(pd.to_numeric, errors="coerce").dropna()

# Check correlation with 'precip_mm'
corr_with_precip = df.corr()["precip_mm"].sort_values(ascending=False)

print(corr_with_precip)
```

**Note:** We see a strong correlation for extreme categorical values

```python
# Drop all columns starting with 'rain_'
merged_df = merged_df.drop(columns=[c for c in merged_df.columns if c.startswith("rain_")])

# Check remaining columns
print("Remaining columns:")
for column in merged_df.columns:
    print(column)

print("\nFinal Dataframe:")
display(merged_df.head())
```
